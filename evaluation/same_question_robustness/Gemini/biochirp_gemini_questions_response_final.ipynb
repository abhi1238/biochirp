```python

import asyncio
import time
import json
import websockets
import requests
import pandas as pd
from pathlib import Path

from gemini_utility import return_gemini_answer

```

```python

model = "gemini-2.5-flash-lite"
dct_result = {}
dct_result[model] = {}

MAX_RETRIES = 10
RUNS = range(1, 6)
```


## Question 1

**Which diseases are associated with BRAF?**


```python

question = 'Which diseases are associated with BRAF?'


dct_result[model].setdefault(question, {})



for run_number in RUNS:
    print(f"\n Starting run {run_number} for query: {question}")

    for attempt in range(1, MAX_RETRIES + 1):
        print(f"  Attempt {attempt}/{MAX_RETRIES}")

        

        try:
            df=None
            latency=None
            start_time = time.perf_counter()
            df = await return_gemini_answer(question)
            latency = time.perf_counter() - start_time

            if latency>5:
                continue

            if len(df.columns)<2:
                continue


            if not isinstance(df, pd.DataFrame):
                raise ValueError("Returned object is not a DataFrame")
            if df.empty:
                raise ValueError("Returned DataFrame is empty")

            dct_result[model][question][run_number] = {
                "rows": df.shape[0],
                "latency": latency,
                "dataframe": df,
            }

            print(
                f" Run {run_number} completed | "
                f"rows={df.shape[0]} | latency={latency:.3f}s"
            )
            break

        except Exception as e:
            latency = time.perf_counter() - start_time
            print(f" Attempt {attempt} failed after {latency:.3f}s: {e}")

            if attempt == MAX_RETRIES:
                print(f" Run {run_number} failed after {MAX_RETRIES} attempts")

    print("-" * 60)
```

```python
import pickle
with open("Gemini_same_question_response_final.pkl", "wb") as f:
    pickle.dump(dct_result, f)

import json
import pandas as pd

def make_json_safe(obj):
    if isinstance(obj, pd.DataFrame):
        return {
            "__type__": "dataframe",
            "columns": obj.columns.tolist(),
            "rows": obj.to_dict(orient="records")
        }
    elif isinstance(obj, dict):
        return {k: make_json_safe(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [make_json_safe(v) for v in obj]
    else:
        return obj


json_safe_result = make_json_safe(dct_result)

with open("Gemini_same_question_response_final.json", "w", encoding="utf-8") as f:
    json.dump(
        json_safe_result,
        f,
        indent=2,
        ensure_ascii=False
    )
```
