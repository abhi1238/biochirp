```python
records = []

for model, queries in dct_result.items():           # model level
    for query, runs in queries.items():             # query level
        for run_id, payload in runs.items():        # run level
            records.append({
                "model": model,
                "query": query,
                "run": int(run_id),
                "rows": int(payload.get("rows", 0)),
                "latency": payload.get("latency", 0)
            })

df_long = pd.DataFrame(records)
```

```python
df_long["model"].value_counts()
```

```python

# df_long already exists with columns:
# ['model', 'query', 'run', 'rows']

heat_df_rows = (
    df_long
    .groupby(["model", "run"])["rows"]
    .median()          # robust to explosions
    .unstack(fill_value=0)
)


heat_df_latency = (
    df_long
    .groupby(["model", "run"])["latency"]
    .median()          # robust to explosions
    .unstack(fill_value=0)
)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

models = list(dct_result.keys())
overlap_matrix = pd.DataFrame(0.0, index=models, columns=models)

def get_clean_row_set(df):
    if not isinstance(df, pd.DataFrame) or df.empty:
        return set()
    
    # FIX 1: Only compare ID columns if they exist, otherwise use all
    # This prevents noise from 'search_rank' or 'score' columns from breaking matches
    id_cols = [c for c in df.columns if 'id' in c.lower() or 'symbol' in c.lower()]
    target_df = df[id_cols] if id_cols else df
    
    # FIX 2: Standardize column order
    target_df = target_df.reindex(sorted(target_df.columns), axis=1)
    
    return set(
        tuple(str(val).strip().lower() if pd.notna(val) else "nan" for val in row)
        for row in target_df.itertuples(index=False, name=None)
    )

# 1. Initialize
row_unions = {model: {} for model in models}

# 2. POPULATE (Move this BEFORE printing)
for model in models:
    for query, runs in dct_result[model].items():
        sets = []
        for run_id, payload in runs.items():
            df = payload.get("dataframe_id")
            row_set = get_clean_row_set(df)
            if row_set:
                sets.append(row_set)
        if sets:
            row_unions[model][query] = set.union(*sets)

# 3. DEBUG PRINT (Now it will have data)
for model in models:
    query_stats = {q: len(v) for q, v in row_unions[model].items()}
    print(f"Model: {model} | Queries Processed: {len(query_stats)} | Total Rows: {sum(query_stats.values())}")

# 4. COMPUTE
for ref_model in models:
    for ret_model in models:
        overlap_sum = 0
        ref_total = 0
        
        common_queries = set(row_unions[ref_model].keys()).intersection(row_unions[ret_model].keys())

        for q in common_queries:
            ref_set = row_unions[ref_model][q]
            ret_set = row_unions[ret_model][q]

            inter_len = len(ref_set.intersection(ret_set))
            overlap_sum += inter_len
            ref_total += len(ref_set)

        if ref_total > 0:
            overlap_matrix.loc[ret_model, ref_model] = overlap_sum / ref_total

# 5. PLOT

plt.figure(figsize=(5, 5))
sns.heatmap(overlap_matrix, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Cross-Model Agreement on Retrieved Biomedical Associations")
plt.xlabel("Reference Model (Denominator)")
plt.ylabel("Retrieved Model (Numerator)")
plt.show()
```

```python
import pandas as pd

dct_jaccard = {}

for model, queries in dct_result.items():
    dct_jaccard[model] = {}
    print(f"\nWorking for model: {model}")

    for question, runs in queries.items():
        run_sets = []

        for run_id, payload in runs.items():
            df = payload.get("dataframe_id")
            

            if not isinstance(df, pd.DataFrame) or df.empty:
                continue

            print(f"Original df: {df.head(1)}")

            # df=process_dataframe_fast(df)

            print(f"Processed df: {df.head(1)}")
            try:
                row_set = set()

                # --- select only *_id columns, sorted ---
                id_cols = sorted([c for c in df.columns if c.endswith("_id")])

                # Skip if no ID columns
                if not id_cols:
                    raise ValueError("No *_id columns found")

                for row in df[id_cols].itertuples(index=False, name=None):
                    cleaned = tuple(
                        str(v).strip().lower()
                        for v in row
                        if v is not None and not pd.isna(v)
                    )

                    # Keep only rows with at least one valid ID
                    if cleaned:
                        row_set.add(cleaned)

                if row_set:
                    run_sets.append(row_set)

            except Exception as e:
                print(f"  ⚠️ Run {run_id} skipped: {e}")


        if len(run_sets) < 2:
            print(f"  ❌ Not enough valid runs for '{question}'")
            dct_jaccard[model][question] = None
            continue

        intersection = set.intersection(*run_sets)
        union = set.union(*run_sets)

        jaccard = len(intersection) / len(union) if union else 0.0
        dct_jaccard[model][question] = jaccard

        print(f"  ✅ Jaccard for '{question}': {jaccard:.3f}")
```

```python
import pandas as pd

records = []

for model, qdict in dct_jaccard.items():
    for question, jac in qdict.items():
        if jac is None:
            continue
        records.append({
            "model": model,
            "query": question,
            "jaccard": jac
        })

df_jaccard = pd.DataFrame(records)
```

```python
order = (
    df_jaccard
    .groupby("model")["jaccard"]
    .median()
    .sort_values(ascending=True)
    .index
)
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5, 5), dpi = 300)

sns.boxplot(
    data=df_jaccard,
    x="model",
    y="jaccard",
    order=order,
    showfliers=True,
    ax=ax
)

sns.stripplot(
    data=df_jaccard,
    x="model",
    y="jaccard",
    order=order,
    color="black",
    alpha=0.6,
    jitter=True,
    size=4,
    ax=ax
)

ax.set_ylabel("Jaccard Stability Index")
ax.set_xlabel("Model")
ax.set_title("Model-wise Stability Across 5 Runs (Jaccard Index)")

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=30,
    ha="right"
)

# plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Set a clean theme and font scale
sns.set_theme(style="whitegrid", context="notebook", font_scale=1.1)

# Define a cohesive palette (optional: use "mako", "rocket", "viridis", etc.)
palette = sns.color_palette("viridis", n_colors=len(order))

fig, ax = plt.subplots(figsize=(6, 6), dpi=300)

# 2. Boxplot: Hide outliers (showfliers=False) and add transparency
sns.boxplot(
    data=df_jaccard,
    x="model",
    y="jaccard",
    order=order,
    showfliers=False,       # Hide outliers here since stripplot shows them
    palette=palette,
    linewidth=1.2,
    width=0.6,              # Slightly narrower boxes look more elegant
    boxprops=dict(alpha=0.7), # Make boxes semi-transparent
    ax=ax
)

# 3. Stripplot: Overlay data points
sns.stripplot(
    data=df_jaccard,
    x="model",
    y="jaccard",
    order=order,
    color="#333333",        # Dark grey is often softer than pure black
    alpha=0.6,
    jitter=0.25,            # Slight jitter to separate points
    size=4,
    linewidth=0,            # Remove white outline around dots for a flatter look
    ax=ax
)

# 4. Polish the Axes
ax.set_ylabel("Jaccard Stability Index", fontweight="bold", labelpad=10)
ax.set_xlabel("Model", fontweight="bold", labelpad=10)
ax.set_title("Model-wise Stability (5 Runs)", fontsize=14, pad=15)

# Rotate labels slightly if needed, but horizontal is best if they fit
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    ha="right"
)

# Remove the top and right borders (spines)
sns.despine(trim=True, left=True)

# Add a subtle grid on the y-axis only (whitegrid theme does this, but we can customize)
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
