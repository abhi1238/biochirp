```python
import pandas as pd
import utility
from io import StringIO
```

```python
df = pd.read_excel("verified_questions.xlsx")
```

```python
with open("../../resources/prompts/llm_table_generator.md", "r", encoding="utf-8") as f:
    prompt_md = f.read()
```

```python
dct_result = dict()
```

```python
number_of_rows = df.shape[0]
```

##### model: gpt-4.1-nano

```python
model="gpt-4.1-nano"
dct_result[model] = dict()

for i in range(number_of_rows):

    question = df.loc[i, "Question"]
    dct_result[model][question] = {}

    for run_number in range(1, 6):
        attempts = 0
        success = False

        while attempts < 3 and not success:
            print(
                f"Currently running run {run_number}, "
                f"attempt {attempts+1} for query {i}: {question}"
            )

            try:
                result = await utility.return_openai_answer(
                    user_prompt=question,
                    system_prompt=prompt_md,
                    model=model
                )

                # ? THIS IS THE KEY FIX
                csv_text = result["answer"]

                df_attempt = pd.read_csv(StringIO(csv_text))

                if isinstance(df_attempt, pd.DataFrame) and not df_attempt.empty:
                    dct_result[model][question][run_number] = result
                    success = True
                else:
                    print("Parsed DataFrame is empty, retrying...")

            except Exception as e:
                print(f"Error occurred: {e}, retrying...")

            attempts += 1

        if not success:
            print(f"Run {run_number} failed after retries, skipping this run.")



```

```python
dct_jacard = dict()
for model in dct_result.keys():
    dct_jacard[model] = dict()

    print(f"Working for model : {model}")
    for question_key in dct_result[model].keys():
        tmp_result = dict()

        for run_number in dct_result[model][question_key].keys():
            try:
                rows_as_lists = pd.read_csv(
                    StringIO(dct_result[model][question_key][run_number]['answer'])
                ).values.tolist()
                # s = set(tuple(row) for row in rows_as_lists)
                # tmp_result[run_number] = set(tuple(sorted(t)) for t in s)
                s = set()
                for row in rows_as_lists:
                    clean_row = tuple(str(v) for v in row if pd.notna(v))
                    if clean_row:
                        s.add(clean_row)

                tmp_result[run_number] = s

            except Exception as e:
                print(f"Error parsing run {run_number} for question {question_key}: {e}")
                # print(rows_as_lists)
                continue

        valid_runs = len(tmp_result)
        expected_runs = 5  # change this if your expected number changes

        # ? Fewer than 2 valid runs => skip Jaccard
        if valid_runs < 2:
            print(f"Not enough valid runs for {question_key} ({valid_runs}/{expected_runs}). Skipping Jaccard.")
            continue

        # ? Some but not all runs valid => warning
        if valid_runs < expected_runs:
            print(f"Warning ? partial runs valid for {question_key}: {valid_runs}/{expected_runs} runs succeeded.")

        sets = list(tmp_result.values())
        intersection = set.intersection(*sets)
        union = set.union(*sets)

        if not union:
            print(f"Empty union for {question_key}, skipping Jaccard.")
            continue
        jaccard_index = len(intersection) / len(union)

        dct_jacard[model][question_key] = jaccard_index
        print(f"Jaccard index for query {question_key}:", jaccard_index)
```
