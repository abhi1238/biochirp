# OpenTargets MCP — Agentic (OpenAI Agents SDK)

This notebook runs each query against multiple models via the OpenAI **Agents SDK** + MCP.
Each model × query is run **twice** (each run in its own cell).


```python
import asyncio
from agents import Agent, Runner
from agents.mcp import MCPServerStreamableHttp
```

```python
MODELS = ['gpt-4.1-nano', 'gpt-4o-mini', 'gpt-4.1-mini', 'gpt-5-nano', 'gpt-5-mini']
MCP_SERVER_LABEL = 'open_targets'
MCP_SERVER_URL = 'https://mcp.platform.opentargets.org/mcp'

QUERY_CASES = [
    (
        'CML complete targets',
        'Provide the complete list of molecular targets associated with chronic myeloid leukemia (CML) in Open Targets.'
    ),
    (
        'CML top targets',
        'Provide the top-ranked molecular targets associated with chronic myeloid leukemia (CML) in Open Targets.'
    ),
    (
        'Top diseases treated with aspirin',
        'List the top diseases treated with aspirin according to Open Targets.'
    ),
    (
        'All diseases treated with aspirin',
        'Provide the complete list of diseases treated with aspirin according to Open Targets.'
    ),
    (
        'Top diseases associated with TP53',
        'List the top diseases associated with the TP53 gene in Open Targets.'
    ),
    (
        'All diseases associated with TP53',
        'Provide the complete list of diseases associated with the TP53 gene in Open Targets.'
    ),
]

SYSTEM_PROMPT = '''You are a deterministic biomedical data retriever for Open Targets MCP.

Rules:
- You MUST use the MCP tools to answer the user question.
- Always retrieve COMPLETE data. If the tool is paginated, fetch all pages before responding.
- Do NOT summarize. Do NOT truncate results.
- Do not ask the user what to do next.
- If no results exist, say so clearly.
'''

# Results store: result[query_key][model][run_number] = final_output
result = {}
```

```python
async def run_agentic_query(model: str, query_text: str) -> str:
    async with MCPServerStreamableHttp(
        name=MCP_SERVER_LABEL,
        params={'url': MCP_SERVER_URL, 'timeout': 60},
    ) as mcp_server:
        agent = Agent(
            name='OpenTargets_MCP_Agent',
            model=model,
            instructions=SYSTEM_PROMPT,
            mcp_servers=[mcp_server],
        )
        result = await Runner.run(agent, query_text)
        return result.final_output
```

### CML complete targets — gpt-4.1-nano — run 1


```python
query_key = 'CML complete targets'
query_text = 'Provide the complete list of molecular targets associated with chronic myeloid leukemia (CML) in Open Targets.'
model = 'gpt-4.1-nano'
run_number = 1

print('\n' + '=' * 80)
print('MODEL:', model)
print('QUERY_KEY:', query_key)
print('RUN_NUMBER:', run_number)
final_output = await run_agentic_query(model, query_text)
result.setdefault(query_key, {}).setdefault(model, {})[run_number] = final_output
print(final_output)
```
